```python
import pandas as pd
articles = pd.read_csv("../data/external/articles.csv")

articles = articles.drop(columns=['length', 'width', 'height', 'weight'])

articles.head()
```

```python
#count proportion of missing values in each column
articles.isnull().sum() / len(articles)
```

```python
import pandas as pd
import numpy as np

audience_to_id = {
    'Dam': '6',
    'Herr': '15',
    'Baby & barn': '12',
    'Barn & ungdom': '42',
    'Generic': '99'
}

articles['audience'] = articles['audience'].fillna('Generic')

def map_audience_to_id(audience_val):
    if pd.isna(audience_val):
        return np.nan
    # Split on commas for multi-valued cells, strip whitespace
    groups = [a.strip() for a in str(audience_val).split(',') if a.strip()]
    # Map each group to its ID
    ids = [audience_to_id[a] for a in groups if a in audience_to_id]
    return ','.join(ids) if ids else np.nan

articles['audienceId'] = articles['audience'].apply(map_audience_to_id)


articles.head()
```

```python
#print unique values of color
color = articles['color'].unique()
print(color)
```

```python
# Clean color and colorId columns by extracting the first unique non-empty token from comma-separated values
import pandas as pd

def clean_first(val):
    if pd.isna(val):
        return pd.NA
    seen = set()
    tokens = []
    for x in str(val).split(','):
        x = x.strip()
        if x and x not in seen:
            seen.add(x)
            tokens.append(x)
    return tokens[0] if tokens else pd.NA

articles['color'] = articles['color'].apply(clean_first)
articles['colorId'] = articles['colorId'].apply(clean_first)

articles['color'] = articles['color'].fillna('__UNK_COLOR__')
articles['colorId'] = articles['colorId'].fillna('__UNK_COLORID__')

articles.head()

```

```python

#  Check if 'sku' and 'groupId' columns are equivalent (i.e., all values are the same row-wise)
equivalent = (articles['sku'] == articles['groupId']).all()
print(f"Are 'sku' and 'groupId' columns equivalent? {equivalent}")

# If not equivalent, show some rows where they differ
if not equivalent:
    diff_rows = articles[articles['sku'] != articles['groupId']]
    print("Rows where 'sku' and 'groupId' differ:")
    display(diff_rows.head())

# Remove the rows where 'sku' and 'groupId' differ, in place
articles = articles[articles['sku'] == articles['groupId']].copy()
```
